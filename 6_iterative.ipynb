```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from pydantic import BaseModel,Field

load_dotenv()
```

```python
generator_llm = ChatGroq(model="gpt-4o", temperature=0.7)
evaluator_llm = ChatGroq(model="gpt-4o", temperature=0.7)
optimizer_llm = ChatGroq(model="gpt-4o", temperature=0.7)
```

```python
class TweetState(TypedDict):
    
    topic: str
    tweet: str
    evaluation: Literal['appropriate', 'inappropriate ']
    iteration: int
    max_iterations: int
```

```python
def generate_tweet(state: TweetState):
    prompt = f"Generate a tweet about {state['topic']} that is appropriate for a general audience."
    
    response = generator_llm.invoke(prompt).content

    return {'tweet': response}
```

```python
def evaluate_tweet(state: TweetState):
    prompt = f"Evaluate the following tweet for appropriateness for a general audience:\n\n{state['tweet']}\n\nRespond with 'appropriate' or 'inappropriate'."
    
    response = evaluator_llm.invoke(prompt).content.strip().lower()

    return {'evaluation': response}
```

```python
graph= StateGraph(TweetState)

graph.add_node('generate_tweet', generate_tweet)
graph.add_node('evaluate_tweet', evaluate_tweet)
graph.add_node('optimize_tweet', optimize_tweet)
```
